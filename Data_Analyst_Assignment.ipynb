#Step:- 7 Assiging the weight slab & generating weight slab

```python
def weight_slab(weight): 
  i = round(weight%1,1)    
  if i == 0.0:
    return weight
  elif i > 0.5:
    return int(weight)+1
  else:
    return int(weight)+0.5
        
```

**generating weight slab**

```python
df_rename['Weight slab'] = df_rename['Total_weight'].apply(weight_slab)
```

```python
df_rename.sample()
```

```python
#applying the same function in df_cc_invoice
df_cc_invoice['weight slab as per shipment'] = df_cc_invoice['Charged Weight'].apply(weight_slab)
df_cc_invoice.head()
```

#Step 8:-computing expected charges:
Courier charges rate card at weight slab and pincode level. If the invoice mentions 
“Forward charges” then only forward charges (“fwd”) should be applicable as per zone 
and fixed & additional weights based on weight slabs. If the invoice mentions “Forward 
and rto charges” then forward charges (“fwd”) and RTO charges (“rto”) should be 
applicable as per zone and fixed & additional weights based on weight slabs

```python
df_rename['Type of Shipment'].value_counts()
```

```python
Total_expected_charge = []
for i in range(len(df_rename)):
  fwd_catogery ='fwd_' + df_rename['Delivery Zone'][i]       #range pases through each delivery zone
  fwd_fixed = df_cc_rates[fwd_catogery+'_fixed'][0]
  fwd_additional = df_cc_rates[fwd_catogery+'_additional'][0]
  rto_catogery = 'rto_' + df_rename['Delivery Zone'][i]
  rto_fixed = df_cc_rates[rto_catogery+'_fixed'][0]
  rto_additional = df_cc_rates[rto_catogery+'_additional'][0]
  if df_rename['Type of Shipment'][i] == 'Forward charges':
    if df_rename['Weight slab'][i]<= 0.5:
      Total_expected_charge.append(fwd_fixed)
    else:
      additional_weight = (df_rename['Weight slab'][i]-0.5)/0.5
      Total_expected_charge.append(fwd_fixed + additional_weight*fwd_additional)
  if df_rename['Type of Shipment'][i] == 'Forward and RTO charges':
    if df_rename['Weight slab'][i]<=0.5:
      Total_expected_charge.append(fwd_fixed + rto_fixed)
    else:
      additional_weigth = (df_rename['Weight slab'][i]-0.5)/0.5
      Total_expected_charge.append(fwd_fixed + additional_weight*(fwd_fixed + rto_fixed))

df_rename['expected_charges(Rs.)'] = Total_expected_charge
```
